 **Xử lí outlier**

```python
def remove_outliers(df, columns):
    for column in columns:
        # Xác định ngưỡng upper và lower để xác định outliers
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Loại bỏ outliers bằng cách thay thế bằng giá trị trung vị của cột
        df[column] = df[column].apply(lambda x: df[column].median() if x < lower_bound or x > upper_bound else x)

# Sử dụng hàm remove_outliers trên DataFrame và chỉ định các cột muốn loại bỏ outliers
remove_outliers(df, ['X3 distance to the nearest MRT station', 'Y house price of unit area','X5 latitude'])
for i in df.describe().columns:
    sns.histplot(df[i],bins=floor(sqrt(len(df[i]))))
    plt.show()
print(df.shape)     
```

## **Model training and evaluation.**


```python
##70%-train,15%-validation,15%-test
X_main, X_test, y_main, y_test = train_test_split(x, y, test_size=0.15,random_state=0)
# Tiếp tục chia tập huấn luyện thành tập huấn luyện và tập validation
X_train, X_val, y_train, y_val = train_test_split(X_main, y_main, test_size=0.176,random_state=0) # 0.176 x 0.8 ~ 0.15


print(f"Training size: \tX_train: {X_train.shape} \ty_train: {y_train.shape}")
print(f"Testing size: \tX_test: {X_test.shape} \ty_test: {y_test.shape}")
print(f"Validation size: \tX_val: {X_val.shape} \ty_val: {y_val.shape}")
```

**Tiến hành scale sau khi đã tách ra train validation và test**

```python
from sklearn.preprocessing import StandardScaler

# Khởi tạo scaler
scaler = StandardScaler()

# Fit scaler trên tập huấn luyện và transform cả tập huấn luyện tập kiểm chứng và tập kiểm tra
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_val_scaled = scaler.transform(X_val)

print(f"Training size: \tX_train: {X_train_scaled.shape} \ty_train: {y_train.shape}")
print(f"Testing size: \tX_test: {X_test_scaled.shape} \ty_test: {y_test.shape}")
print(f"Validation size: X_val: {X_val_scaled.shape} \ty_val: {y_val.shape}")
```

**Tiếp theo sẽ viết 1 hàm chọn bậc của polynomialFeatures, hàm sẽ hỗ trợ vẽ plot và cho thấy global min của Validation set tại đó ta có thể quyết định chọn bậc**

**Cụ thể Polynomial sẽ tạo ra các biến đa thức từ các biến độc lập của chúng ta**

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
def choose_degree(model,maxdegree):
    train_errors = []
    val_errors = []

    for degree in range(1, maxdegree + 1):
        # Tạo các biến đa thức từ dữ liệu huấn luyện và validation
        poly = PolynomialFeatures(degree=degree, include_bias=False)
        X_poly_train = poly.fit_transform(X_train_scaled)
        X_poly_val = poly.transform(X_val_scaled)
        
        # Huấn luyện mô hình với các biến đa thức
        model.fit(X_poly_train, y_train)
        
        # Đánh giá hiệu suất trên tập huấn luyện và tập validation
        y_pred_train = model.predict(X_poly_train)
        y_pred_val = model.predict(X_poly_val)
        
        train_error = mean_squared_error(y_train, y_pred_train)
        val_error = mean_squared_error(y_val, y_pred_val)
        
        train_errors.append(train_error)
        val_errors.append(val_error)

    # Vẽ đồ thị hiệu suất trên tập huấn luyện và tập validation theo bậc của biến đa thức
    plt.plot(range(1, maxdegree + 1), train_errors, label='Training Error')
    plt.plot(range(1, maxdegree + 1), val_errors, label='Validation Error')
    plt.xlabel('Degree of Polynomial')
    plt.ylabel('Mean Squared Error')
    plt.title('Training and Validation Error vs. Polynomial Degree')
    plt.legend()
    plt.show()
```

```python
from sklearn.metrics import r2_score
from sklearn import linear_model
linear_regression = linear_model.LinearRegression()
choose_degree(linear_regression,3)
```

**Có nhiều trường hợp random bậc min của validation set tương đối lớn nhưng nếu ta chọn bậc quá lớn có thể bị overfit nên ta chỉ chọn 2**

## **Hyperparameter Tuning.**

**Thử training với Ridge Regression (LNR nhưng có thêm Regularization)**

**Tìm alpha tối ưu khi chạy mô hình**

```python
from sklearn.model_selection import GridSearchCV
# Phân chia dữ liệu thành tập huấn luyện và tập kiểm tra
# Tạo một đối tượng Lasso Regression
Rid_regress=linear_model.Ridge()

# Thiết lập các giá trị alpha muốn thử nghiệm
alphas = [0.00015,0.0015,0.015,0.15, 1, 10, 100]

# Tạo một grid search để tìm ra alpha tối ưu
param_grid = {'alpha': alphas}
grid_search = GridSearchCV(Rid_regress, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_scaled,y_train)
best_alpha = grid_search.best_params_['alpha']

# In ra giá trị alpha tối ưu và điểm số tương ứng
print("Best alpha for Ridge:", best_alpha)
print("Best MSE:", -grid_search.best_score_)
```

**Lí do sử dụng neg_MSE là do trong gridsearch các metric được coi là càng lớn càng tốt nên ta sẽ cho MSE âm để khi nó trả về giá trị âm lớn nhất thì đó sẽ là giá trị dương nhỏ nhất của MSE (MSE càng nhỏ thì càng tốt)**

```python
best_Ridge=linear_model.Ridge(best_alpha)
choose_degree(best_Ridge,4)
```
